```python
def download(filename, url):
    # We do not download again if the file already exists
    if os.path.exists(filename):
        return (str(filename) + " already exists!")

    # TODO: Write the code to download the file from URL
    r = requests.get(url)
    r.raise_for_status()
    data = r.text
    f = open(filename, "w", encoding="utf-8")
    f.write(data)
    f.close()
#     f = open("hello.txt", encoding="utf-8")
#     data = f.read()
#     f.close()
    return (str(filename) + " created!")
```

```python
h_trs = doc.find('tr')
h_list = []

for h in h_trs.find_all('td'):
    h_list.append(h.get_text())
    
h_list
```

```python
c_trs = doc.find_all('tr')[1:]
country_list = []
for tr in c_trs:
    c_list = []
    for td in tr.find_all('td'):
        c_list.append(td.get_text())
    country_list.append(c_list)

country_list
```

```python
country_dictList = []
for each in country_list:
    row = dict()
    for i in range(len(each)-2):
        row[h_list[i]] = each[i]
    country_dictList.append(row)
```

```python
# Q13: What is the distance between Camp Randall Stadium and the Wisconsin State Capital?
crs = (43.070231, -89.411893) # (latitude1, logitude1)
wsc = (43.074645, -89.384113) # (latitude2, logitude2)
radius = 3956

dlon = -89.411893 - (-89.384113)
dlat = 43.070231 - 43.074645

a = math.sin(math.radians(dlat/2))*math.sin(math.radians(dlat/2))+math.cos(math.radians(crs[0]))*math.cos(math.radians(wsc[0]))*math.sin(math.radians(dlon/2))*math.sin(math.radians(dlon/2))
c = 2 * math.asin((min(1,math.sqrt(a))))
d = radius * c
d
```

```python
# Q14: What is the distance between France and Japan?
france_location = ()
japan_location = ()
for index in capitals.index:
    if capitals["Country Name"][index] == "France":
        lat = capitals["Capital Latitude"][index]
        log = capitals["Capital Longitude"][index]
        france_location = (lat, log)
    if capitals["Country Name"][index] == "Japan":
        lat = capitals["Capital Latitude"][index]
        log = capitals["Capital Longitude"][index]
        japan_location = (lat, log)
        
fl = france_location # (latitude1, logitude1)
jl = japan_location # (latitude2, logitude2)
radius = 3956

dlo = japan_location[1] - france_location[1]
dla = japan_location[0] - france_location[0]

Q14_a = math.sin(math.radians(dla/2))*math.sin(math.radians(dla/2))+math.cos(math.radians(fl[0]))*math.cos(math.radians(jl[0]))*math.sin(math.radians(dlo/2))*math.sin(math.radians(dlo/2))
Q14_c = 2 * math.asin((min(1,math.sqrt(Q14_a))))
Q14_d = radius * Q14_c
Q14_d
```

```python
def distance(loc1, loc2):
    radius = 3956

    dlo = loc1[1] - loc2[1]
    dla = loc1[0] - loc2[0]

    a = math.sin(math.radians(dla/2))*math.sin(math.radians(dla/2))+math.cos(math.radians(fl[0]))*math.cos(math.radians(jl[0]))*math.sin(math.radians(dlo/2))*math.sin(math.radians(dlo/2))
    c = 2 * math.asin((min(1,math.sqrt(a))))
    d = radius * c
    
    return d
```
